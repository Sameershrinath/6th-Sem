## Standard deviation

Standard Deviation (SD) is a statistical measure that tells us how much the data values deviate from the mean.

Low SD → data points are close to the mean

High SD → data points are widely spread

It is one of the most important concepts in statistics, probability, and machine learning.

Mean

$$
\bar{x} = \frac{1}{n}\sum_{i=1}^{n} x_i
$$


Variance

$$
\sigma^2 = \frac{1}{n}\sum_{i=1}^{n}(x_i - \bar{x})^2
$$


standard deviation

$$
\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^{n}(x_i - \bar{x})^2}
$$


```python
def variance(data):
    n=len(data)

    mean=sum(data)/n
    variance=0
    for i in data:
        variance+=(i-mean)**2
    variance=variance/n
    return variance
```

```python
import math

def standard_deviation(data):
    n=len(data)

    mean=sum(data)/n
    variance=0
    for i in data:
        variance+=(i-mean)**2
    variance=variance/n

    std_dev=math.sqrt(variance)
    return std_dev
```

```python
data = [2, 4, 6, 8]
standard_deviation(data)
```
